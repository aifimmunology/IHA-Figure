```r
library(arrow)
library(flowCore)
library(MetaCyto)
library(tidyverse)
library(data.table)
library(ggplot2)
library(FlowSOM)
library(uwot)
library(patchwork)
library(reshape2)
library(gridExtra)
library(parallel)
library(ggpubr)
library(pheatmap)
set.seed(42)
```

```r
df<-arrow::read_parquet('clustered.parquet')
```

```r
df_subset<-df[sample(nrow(df), 100000), ]
ggplot(df_subset, aes(x = `CD4`, y = `CD8`)) +
    geom_point(color = "black", pch = 21) +
    theme(legend.position = "right") +
    theme_minimal()
```

```r
df<-df %>% 
  mutate(L1 = case_when(
    CD4 >= 5 & CD8 < 4 ~ "CD4+",
    CD4 < 5 & CD8 >= 4 ~ "CD8+",
    CD4 >= 5 & CD8 >= 4 ~ "CD4+CD8+",
    CD4 < 5 & CD8 < 4 ~ "CD4-CD8-"
  ))
```

```r
df_CD4<-df %>% filter(L1=="CD4+")
df_CD8<-df %>% filter(L1=="CD8+")
```

# CD4

```r
df_subset <- df_CD4[sample(nrow(df_CD4), 10000), ]
ggplot(df_subset, aes(y = CD45RA, x = CCR7)) +
    geom_point(color = "black", pch = 21) +
    theme(legend.position = "right") +
    theme_minimal() +
    geom_vline(xintercept = 3) +
    geom_hline(yintercept = 5.5)
```

```r
df_CD4<-df_CD4 %>% 
  mutate(L2 = case_when(
    CCR7 > 3 & CD45RA > 5.5 ~ "Naive",
    TRUE~ "Memory"

  ))
```

```r
CD4_Memory<-df_CD4 %>% filter(L2=="Memory")
CD4_Naive<-df_CD4 %>% filter(L2=="Naive")
```

```r
df_subset<-CD4_Memory[sample(nrow(CD4_Memory),10000), ]
ggplot(df_subset, aes(y = `CD27`, x = `CCR7`)) +
    geom_point(color = "black", pch = 21) +
    theme(legend.position = "right") +
    theme_minimal()+ geom_vline(xintercept = 2.5)+ geom_hline(yintercept = 3.5)
```

```r
CD4_Memory<-CD4_Memory %>% 
  mutate(L3 = case_when(
    CCR7 < 2.5 & CD27 >3.5 ~ "CD27+ EM",
    CCR7 > 2.5 & CD27 >3.5 ~ "CD27+ CM",
    CCR7 < 2.5 & CD27 <3.5 ~ "CD27- EM",  
    TRUE~"Other Memory"

  ))
```

```r
df_subset<-CD4_Memory[sample(nrow(CD4_Memory),30000), ]
ggplot(df_subset, aes(x = `IL4`, y = `IFNg`)) +
    geom_point(color = "black", pch = 21) +
    theme(legend.position = "right") +
    theme_minimal()+ geom_vline(xintercept = 3.3)+ geom_hline(yintercept = 3)
```

```r
CD4_Memory<-CD4_Memory %>% 
  mutate(L4 = case_when(
      IL4 >3.3 & IFNg >3 ~ "IFNg+ IL4+",
      IL4 >3.3 & IFNg <3 ~ "IFNg- IL4+",
      IL4 <3.3 & IFNg >3 ~ "IFNg+ IL4-",
      TRUE~ "IFNg- IL4-"

  ))
```

```r
CD4<-plyr::rbind.fill(CD4_Memory,CD4_Naive)
```

```r
CD4$Final_Celltype<-paste(CD4$L4,CD4$L3,CD4$L2,CD4$L1)
unique(CD4$Final_Celltype)
```
