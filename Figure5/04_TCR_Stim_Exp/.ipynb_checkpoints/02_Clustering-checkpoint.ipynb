```r
library(arrow)
library(flowCore)
library(MetaCyto)
library(tidyverse)
library(data.table)
library(ggplot2)
library(FlowSOM)
library(uwot)
library(patchwork)
library(reshape2)
library(gridExtra)
library(parallel)
library(ggpubr)
library(pheatmap)
set.seed(42)
```

```r
df<-as.data.frame(read_parquet('filtered.parquet'))
```

```r
colnames(df)
```

```r
Target <- c("CD3", "CD45RA", "CD27", "CXCR5", "CD8","CD4", "IL4",'IFNg',"CD40L", "CCR7")
Target <- c("CD3", "CD45RA", "CD27", "CXCR5", "CD8","CD4",  "CCR7")

#Target <- c("CD3", "CD45RA", "CD27", "CXCR5", "CD8","CD4",, "CD40L", "CCR7")
```

```r
#df_subset<-df[sample(nrow(df), 500000), ]
```

```r
df_subset<-df
```

```r
clustering_res<-FlowSOM::SOM(as.matrix(df_subset[Target]),
                         xdim = 5,  
                         ydim = 5)
```

```r
meta_clusters <- FlowSOM::metaClustering_consensus(clustering_res$codes, k = 10)
```

```r
df_subset$clustering<-clustering_res$mapping[,1]
```

```r
uamp_res<-umap(as.matrix(df_subset[Target]), n_neighbors=25, min_dist=0.3, n_threads =25)
```

```r
df_subset$UMAP_1<-uamp_res[,1]
df_subset$UMAP_2<-uamp_res[,2]
```

```r

options(repr.plot.width = 20, repr.plot.height =10)

columns_to_plot <- c(Target,c("IL4","FVS510"))

plots <- lapply(columns_to_plot, function(col) {
  ggplot(df_subset, aes(x=UMAP_1, y=UMAP_2)) + 
    geom_point(aes_string(color=col), shape=16,size=0.00001) + 
    ggtitle(col) +
    theme_minimal()+
  scale_colour_gradient2()
})

ggarrange(plotlist = plots, ncol = 4, nrow = 4)
```

```r
options(repr.plot.width = 8, repr.plot.height =8)
ggplot(df_subset, aes(x=UMAP_1, y=UMAP_2)) + 
    geom_point(aes(color=as.character(clustering)), shape=16,size=0.001) + 
    theme_minimal()
```

```r
centroids <- df_subset %>%
  group_by(clustering) %>%
  summarise(UMAP_1 = mean(UMAP_1, na.rm = TRUE), 
            UMAP_2 = mean(UMAP_2, na.rm = TRUE))

# Plot with labels at cluster centroids
options(repr.plot.width = 8, repr.plot.height = 8)
ggplot(df_subset, aes(x = UMAP_1, y = UMAP_2)) + 
  geom_point(aes(color = as.character(clustering)), shape = 16, size = 0.001) + 
  geom_text(data = centroids, aes(label = clustering), size = 4, fontface = "bold", color = "black") +  # Add labels
  theme_minimal()
```

```r
options(repr.plot.width = 8, repr.plot.height =8)
ggplot(df_subset, aes(x=UMAP_1, y=UMAP_2)) + 
    geom_point(aes(color=as.character(clustering)), shape=16,size=0.001) + 
    theme_minimal()
```

```r
df_subset_summary<-df_subset%>% 
  group_by(clustering) %>% 
  summarise(across(1:17, mean, na.rm = TRUE))
colnames(df_subset_summary)[-1] <- colnames(df_subset)[1:18]  # Assign proper column names

# Convert clustering column to row names
df_subset_summary <- column_to_rownames(df_subset_summary, var = "clustering")
```

```r


pheatmap(t(as.matrix(df_subset_summary[7:17])),
         scale = "row", 
         clustering_rows = TRUE, 
         clustering_cols = FALSE,breaks=seq(-2, 2, length.out = 100), 
         color = colorRampPalette(c("blue", "white", "red"))(100),
         main = "Heatmap of Mean Values (Grouped by Clustering)")
```
