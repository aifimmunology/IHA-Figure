# Read FreqChanges and DEG number 

```r
df<-read.csv("Plotting_Input_Files/FreqChanges_and_DEG.csv")
```

```r
labels_vector<-c("Core naive CD8 T cell","CM CD8 T cell","GZMK+ CD27+ EM CD8 T cell","Naive CD4 Treg",
                 "Core naive CD4 T cell","CM CD4 T cell","GZMB- CD27- EM CD4 T cell","GZMB- CD27+ EM CD4 T cell")
```

```r
options(repr.plot.width = 4, repr.plot.height = 4)

ggplot(df, aes(x=delta_change, y=DEG_Counts, size=adjP_degs_combined_value, fill=AIFI_L3)) +
  geom_point(shape=21, colour="black") +
  geom_text_repel(data=df%>% filter(AIFI_L3 %in% labels_vector),
                  aes(label=AIFI_L3), 
                  size=3, nudge_y=10) +
  theme_bw(base_size=10) +
  scale_size(range = c(0.1, 10)) +
  xlab("Delta_Freq-CLR") +
  theme(legend.position="bottom") +
  theme(legend.position = "none") +
  scale_fill_manual(values=color_vector) +
  scale_x_continuous(limits = c(-max(abs(df$delta_change)), max(abs(df$delta_change))))
```
