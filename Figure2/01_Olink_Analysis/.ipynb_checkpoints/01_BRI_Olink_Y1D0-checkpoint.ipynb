```r
Olink<-Olink %>% filter(sample.sampleKitGuid %in% unique(meta$sample.sampleKitGuid))
```

```r
Olink<-left_join(Olink,meta, by = c("sample.sampleKitGuid"))
```

```r
res_list<-mclapply(unique(Olink$Assay),function(x){
Olnik_sub<-Olink  %>% filter(sample.visitName=='Flu Year 1 Day 0',Assay==x)
Olnik_sub$NPX_bridged<-as.numeric(Olnik_sub$NPX_bridged)
Olnik_sub[!is.na(Olnik_sub$NPX_bridged),]
res<-lm(NPX_bridged ~ cohort.cohortGuid+CMV+subject.biologicalSex, data=Olnik_sub)
res<-as.data.frame(summary(res)$coefficients)[2:4,]
res$factor<-c('Age','CMV','Sex')  
res$Assay<-x
    
res$Log2FC_Age<-Olnik_sub %>%
  dplyr::group_by(Assay, cohort.cohortGuid) %>%
  dplyr::summarise(Mean_Expression = mean(NPX_bridged, na.rm = TRUE)) %>%
  spread(cohort.cohortGuid, Mean_Expression) %>%
  mutate(Log2FC = log2(BR2 / BR1))%>% ungroup() %>% select(Log2FC) %>% pull()
res$delta_changes_Ages<-Olnik_sub %>%
  dplyr::group_by(Assay, cohort.cohortGuid) %>%
  dplyr::summarise(Mean_Expression = mean(NPX_bridged, na.rm = TRUE)) %>%
  spread(cohort.cohortGuid, Mean_Expression) %>%
  mutate(diff = BR2- BR1)%>% select(diff) %>% pull()   
    

res$Log2FC_CMV<-Olnik_sub %>%
  dplyr::group_by(Assay, CMV) %>%
  dplyr::summarise(Mean_Expression = mean(NPX_bridged, na.rm = TRUE)) %>%
  spread(CMV, Mean_Expression) %>%
  mutate(Log2FC = log2(Positive / Negative))%>% ungroup() %>% select(Log2FC) %>% pull()
res$delta_changes_CMV<-Olnik_sub %>%
  dplyr::group_by(Assay, CMV) %>%
  dplyr::summarise(Mean_Expression = mean(NPX_bridged, na.rm = TRUE)) %>%
  spread(CMV, Mean_Expression) %>%
  mutate(diff = Positive- Negative)%>% select(diff) %>% pull()    
    

res$Log2FC_Sex<-Olnik_sub %>%
  dplyr::group_by(Assay, subject.biologicalSex) %>%
  dplyr::summarise(Mean_Expression = mean(NPX_bridged, na.rm = TRUE)) %>%
  spread(subject.biologicalSex, Mean_Expression) %>%
  mutate(Log2FC = log2(Female / Male))%>% ungroup() %>% select(Log2FC) %>% pull()
res$delta_changes_Sex<-Olnik_sub %>%
  dplyr::group_by(Assay, subject.biologicalSex) %>%
  dplyr::summarise(Mean_Expression = mean(NPX_bridged, na.rm = TRUE)) %>%
  spread(subject.biologicalSex, Mean_Expression) %>%
  mutate(diff = Female- Male)%>% select(diff) %>% pull()    
    
    
    
return(res)
},mc.cores=60)
```

```r
res_all<-do.call(rbind,res_list)
res_all<-res_all %>%group_by(factor) %>% mutate(padj=p.adjust(`Pr(>|t|)`))
write.csv(res_all,"Olink_lm_Result.csv")
```

### Ages

```r
res_all_single_factor<-res_all %>% filter(factor=='Age')
```

```r
data <- res_all_single_factor%>% mutate(Expression = case_when(delta_changes_Ages >=0.1 & padj <= 0.05 ~ "Up-Older",
                                                     delta_changes_Ages <= -0.1 & padj <= 0.05 ~ "Up-Young",
                                                     TRUE ~ "Unchanged"))

top <- 30
top_genes <- bind_rows(
  data %>% 
    filter(Expression == 'Up-Older') %>% 
    arrange(padj, desc(abs(delta_changes_Ages))) %>% 
    head(top),
  data %>% 
    filter(Expression == 'Up-Young') %>% 
    arrange(padj, desc(abs(delta_changes_Ages))) %>% 
    head(top))
```

```r
options(repr.plot.width = 5, repr.plot.height = 3)

p_age<-ggplot(data, aes(delta_changes_Ages, -log(padj,10))) +
  geom_point(aes(color = Expression), size = 0.1) +
  xlab(expression("Delta Changes")) + 
  ylab(expression("-log"[10]*"FDR")) +
  scale_color_manual(values = c("Up-Older"='#bf812d', "Unchanged"="gray50", "Up-Young"='#35978f')) +
  guides(colour = guide_legend(override.aes = list(size=1.5)),alpha = "none") +
  geom_text_repel(data = top_genes,box.padding = 0.5,force= 0.5,max.overlaps=50,
    nudge_x           = 0.15,
    hjust             = 0.1,
    segment.size      = 0.1,
                   mapping = aes(delta_changes_Ages, -log(padj,10), label = Assay),
                   size = 2)+theme_classic(base_size=6)+ theme(legend.position="none")+theme_bw(base_size=10)
p_age
```
