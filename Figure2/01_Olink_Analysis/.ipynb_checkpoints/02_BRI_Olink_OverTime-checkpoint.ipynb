# Plotting test

```r
Olink_single<-Olink %>% filter(Assay=='CXCL17')
```

```r
Olink_single <- Olink_single %>% dplyr::mutate(`Age Group` = case_when(
  cohort.cohortGuid == "BR1" ~ "Young",
  cohort.cohortGuid == "BR2" ~ "Older",
  TRUE ~ NA_character_
))
Olink_single$`Age Group`<-factor(Olink_single$`Age Group`,levels=c("Young","Older"))
```

```r
options(repr.plot.width = 60/25.4, repr.plot.height = 60/25.4)

p <- ggplot(Olink_single %>% filter(sample.daysSinceFirstVisit<600), 
            aes(sample.daysSinceFirstVisit, NPX_bridged,color=`Age Group`)) +
    geom_smooth(method="lm",se=TRUE, linewidth=0.2) + 
scale_color_manual(values = c("Older"='#bf812d',  "Young"='#35978f'))+
    theme_bw(base_size = 6) +
    theme(axis.line = element_line(colour = "black", linewidth = 0.1),
          legend.position = "top",
          legend.title = element_blank(),
          panel.grid.major = element_blank(),
          panel.grid.minor = element_blank(),
          panel.border = element_blank(),
          panel.background = element_blank(),
          strip.background = element_rect(linewidth = 0),
          axis.title.x = element_text(margin = margin(t = 2)))+
geom_line(aes(group = interaction(`Age Group`, subject.subjectGuid)), 
              linewidth=0.1, alpha=0.5,
              linetype =1)
p
```

```r
options(repr.plot.width = 80/25.4, repr.plot.height = 60/25.4)

density_plot <- ggplot(Olink_single %>% filter(sample.daysSinceFirstVisit < 600), 
                       aes(NPX_bridged, fill = `Age Group`, color = `Age Group`)) +
    geom_density(alpha = 0.5) +
    scale_fill_manual(values = c("Older" = '#bf812d', "Young" = '#35978f')) +
    scale_color_manual(values = c("Older" = '#bf812d', "Young" = '#35978f')) +
    coord_flip() +
    theme_void() +
    theme(legend.position = "none")

# Combine the plots
combined_plot <- plot_grid(p, density_plot, ncol = 2, rel_widths = c(10, 1), align = 'h', axis = 'tb')

print(combined_plot)
```
