```r
library(tidyverse)
library(data.table)
library(ggplot2)
library(patchwork)
library(reshape2)
set.seed(42)
```

```r
  get_density <- function(x, y, ...) {
    dens <- MASS::kde2d(x, y, ...)
    ix <- findInterval(x, dens$x)
    iy <- findInterval(y, dens$y)
    ii <- cbind(ix, iy)
    return(dens$z[ii])
  }  
```

```r
df<-arrow::read_parquet('../Figure5/04_TCR_Stim_Exp/clustered.parquet')
```

```r
plot_quadrant_density <- function(data, xvar, yvar, x_cut, y_cut,
                                  sample_size = 3e5, 
                                  density_n = 100,
                                  xlabels="",
                                  ylabels="",
                                  width = 5, height = 5) {
  xq <- rlang::enquo(xvar)
  yq <- rlang::enquo(yvar)

  df_sub <- data %>%
    dplyr::sample_n(size = min(sample_size, dplyr::n()), replace = FALSE) %>%
    dplyr::mutate(density = get_density(!!xq, !!yq, n = density_n),
                  region = dplyr::case_when(
                    !!xq >= x_cut & !!yq >= y_cut ~ "Q1",
                    !!xq <  x_cut & !!yq >= y_cut ~ "Q2",
                    !!xq <  x_cut & !!yq <  y_cut ~ "Q3",
                    !!xq >= x_cut & !!yq <  y_cut ~ "Q4"
                  ))

  region_pct <- df_sub %>%
    dplyr::count(region) %>%
    dplyr::mutate(percent = round(n / sum(n) * 100, 2))

  xr <- range(dplyr::pull(df_sub, !!xq))
  yr <- range(dplyr::pull(df_sub, !!yq))

  corners <- tibble::tibble(
    region = c("Q1","Q2","Q3","Q4"),
    x      = c(xr[2], xr[1], xr[1], xr[2]),
    y      = c(yr[2], yr[2], yr[1], yr[1])
  )

  region_pct <- region_pct %>%
    dplyr::left_join(corners, by = "region") %>%
    dplyr::mutate(
      offx = 0.08 * diff(xr),
      offy = 0.08 * diff(yr),
      x    = ifelse(region %in% c("Q2","Q3"), x + offx, x - offx),
      y    = ifelse(region %in% c("Q3","Q4"), y + offy, y - offy)
    ) %>%
    dplyr::select(region, percent, x, y)

  options(repr.plot.width = width, repr.plot.height = height)

  ggplot2::ggplot(df_sub, ggplot2::aes(x = !!xq, y = !!yq)) +
    ggplot2::theme_bw(base_size=6) +
    ggplot2::theme(legend.position = "none")+
    ggrastr::geom_point_rast(ggplot2::aes(color = density),
                             stroke = 0, size = 0.01) +
    viridis::scale_color_viridis(option = "H", direction = 1, guide = FALSE) +
    ggplot2::geom_text(
      data = region_pct,
      ggplot2::aes(x = x, y = y, label = paste0(percent, "%")),
      inherit.aes = FALSE, size = 2.1
    )+xlab(xlabels)+ylab(ylabels) +
    ggplot2::geom_vline(xintercept = x_cut,linewidth=0.2) +
    ggplot2::geom_hline(yintercept = y_cut,linewidth=0.2) 
}
```

```r

p <- plot_quadrant_density(
  data        = df,
  xvar        = CD4,
  yvar        = CD8,
  x_cut       = 5,
  y_cut       = 4,
  sample_size = 3e5,
  density_n   = 100,
    xlabels="RB613::CD4",
    ylabels="Spark Blue 550::CD8"

)
print(p)

ggsave("FigureA1.pdf",
       p,
       width=30/25.4,
       height = 30/25.4)
```

```r
df<-df %>% 
  mutate(L1 = case_when(
    CD4 >= 5 & CD8 < 4 ~ "CD4+",
    CD4 < 5 & CD8 >= 4 ~ "CD8+",
    CD4 >= 5 & CD8 >= 4 ~ "CD4+CD8+",
    CD4 < 5 & CD8 < 4 ~ "CD4-CD8-"
  ))
```

```r
df_CD4<-df %>% filter(L1=="CD4+")
df_CD8<-df %>% filter(L1=="CD8+")
```

# CD4

```r
p <- plot_quadrant_density(
  data        = df_CD4,
  xvar        = CCR7,
  yvar        = CD45RA,
  x_cut       = 3,
  y_cut       = 5.5,
  sample_size = 3e5,
  density_n   = 100,
    xlabels="APC-Fire 750::CCR7",
    ylabels="BUV496::CD45RA"

)
print(p)

ggsave("FigureA2.pdf",
       p,
       width=30/25.4,
       height = 30/25.4)
```

```r
df_CD4<-df_CD4 %>% 
  mutate(L2 = case_when(
    CCR7 > 3 & CD45RA > 5.5 ~ "Naive",
    TRUE~ "Memory"

  ))
```

```r
CD4_Memory<-df_CD4 %>% filter(L2=="Memory")
CD4_Naive<-df_CD4 %>% filter(L2=="Naive")
```

```r
p <- plot_quadrant_density(
  data        = CD4_Memory,
  xvar        = CCR7,
  yvar        = CD27,
  x_cut       = 2.5,
  y_cut       = 3.5,
  sample_size = 3e5,
  density_n   = 100,
    xlabels="APC-Fire 750::CCR7",
    ylabels="BUV805::CD27"

)
print(p)

ggsave("FigureA3.pdf",
       p,
       width=30/25.4,
       height = 30/25.4)
```

```r
df_subset<-CD4_Memory[sample(nrow(CD4_Memory),10000), ]
ggplot(df_subset, aes(y = `CD27`, x = `CCR7`)) +
    geom_point(color = "black", pch = 21) +
    theme(legend.position = "right") +
    theme_minimal()+ geom_vline(xintercept = 2.5)+ geom_hline(yintercept = 3.5)
```

```r
CD4_Memory<-CD4_Memory %>% 
  mutate(L3 = case_when(
    CCR7 < 2.5 & CD27 >3.5 ~ "CD27+ EM",
    CCR7 > 2.5 & CD27 >3.5 ~ "CD27+ CM",
    CCR7 < 2.5 & CD27 <3.5 ~ "CD27- EM",  
    TRUE~"Other Memory"

  ))
```

```r
p <- plot_quadrant_density(
  data        = CD4_Memory,
  xvar        = IL4,
  yvar        = IFNg,
  x_cut       = 3.3,
  y_cut       = 3,
  sample_size = 3e5,
  density_n   = 100,
    xlabels="BV421::IL4",
    ylabels="Alexa Fluor 647::IFNg"

)
print(p)

ggsave("FigureA4.pdf",
       p,
       width=30/25.4,
       height = 30/25.4)
```

```r
df_subset<-CD4_Memory[sample(nrow(CD4_Memory),30000), ]
ggplot(df_subset, aes(x = `IL4`, y = `IFNg`)) +
    geom_point(color = "black", pch = 21) +
    theme(legend.position = "right") +
    theme_minimal()+ geom_vline(xintercept = 3.3)+ geom_hline(yintercept = 3)
```

```r
CD4_Memory<-CD4_Memory %>% 
  mutate(L4 = case_when(
      IL4 >3.3 & IFNg >3 ~ "IFNg+ IL4+",
      IL4 >3.3 & IFNg <3 ~ "IFNg- IL4+",
      IL4 <3.3 & IFNg >3 ~ "IFNg+ IL4-",
      TRUE~ "IFNg- IL4-"

  ))
```

```r
CD4<-plyr::rbind.fill(CD4_Memory,CD4_Naive)
```

```r
CD4$Final_Celltype<-paste(CD4$L4,CD4$L3,CD4$L2,CD4$L1)
unique(CD4$Final_Celltype)
```
