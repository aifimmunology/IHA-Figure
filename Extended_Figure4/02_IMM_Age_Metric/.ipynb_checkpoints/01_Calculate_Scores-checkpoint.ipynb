# Get which cell type are changing with Age

```r
BRI_Freq_Data<-read.csv("/home//workspace/private/bri_figure_all_files_test/jupyter/BRI_Figures_Final_V2/Dataset/diha_AIFI_L3_frequencies_2024-05-05.csv")
```

```r
BRI_Freq_Data<-BRI_Freq_Data %>% group_by(sample.sampleKitGuid) %>%
  dplyr::mutate(percentage = (AIFI_L3_count / sum(AIFI_L3_count)) * 100) %>% ungroup()
```

```r
BRI_Freq_Data_subset<-BRI_Freq_Data %>% filter(sample.visitName=="Flu Year 1 Day 0")
```

```r
#compute freq changes
freq_changes<-BRI_Freq_Data_subset %>%
  select(percentage, AIFI_L3_clr,AIFI_L3, cohort.cohortGuid, subject.biologicalSex, subject.cmv) %>%
  group_by(AIFI_L3, cohort.cohortGuid) %>%
  summarise(mean_AIFI_L3_clr = median(AIFI_L3_clr, na.rm = TRUE)) %>%
  spread(cohort.cohortGuid, mean_AIFI_L3_clr) %>%
  mutate(log2fc = log2(`BR2` / `BR1`)) %>%
  mutate(delta_change = `BR2` - `BR1`) 

#compute freq changes p values

freq_changes_STAT<-BRI_Freq_Data_subset %>%
  select(AIFI_L3_clr,AIFI_L3, cohort.cohortGuid,subject.biologicalSex,subject.cmv) %>%
  group_by(AIFI_L3)  %>%
  wilcox_test(AIFI_L3_clr ~ cohort.cohortGuid) %>%
  adjust_pvalue(method='BH') %>%
  add_significance() %>% as.data.frame()

df<-left_join(freq_changes_STAT,freq_changes)
```

```r
df_subset<-df %>% filter(p.adj<0.05)
```

```r
df_subset
```

# Scaling

```r
scale_frequencies <- function(frequencies) {
  lower <- quantile(frequencies, 0.1)
  upper <- quantile(frequencies, 0.9)
  filtered <- frequencies[frequencies >= lower & frequencies <= upper]
  mean_val <- mean(filtered)
  sd_val <- sd(filtered)
  
  scaled <- (frequencies - mean_val) / sd_val
  return(scaled)
}
```

```r
scaled_data <- BRI_Freq_Data_subset %>%
  group_by(AIFI_L3) %>%
  mutate(ScaledFrequency = scale_frequencies(percentage)) %>%
  ungroup()
```

# PCA

```r
wide_data <- scaled_data %>%
  select(sample.sampleKitGuid, AIFI_L3, ScaledFrequency) %>%
  pivot_wider(names_from = AIFI_L3, values_from = ScaledFrequency) %>%
  column_to_rownames(var = "sample.sampleKitGuid") 
```

```r
wide_data<-wide_data[df_subset$AIFI_L3]
```

```r
wide_data[is.na(wide_data)] <- 0
```

```r
pca_result <- prcomp(wide_data, center = TRUE, scale. = TRUE)

summary(pca_result)

pca_data <- as.data.frame(pca_result$x)
pca_data$SampleID <- rownames(pca_data)
```

```r
pca_data<-left_join(pca_data,unique(scaled_data %>% select(cohort.cohortGuid,sample.sampleKitGuid,sample.visitName
)),by=c("SampleID"="sample.sampleKitGuid"))
```

```r
options(repr.plot.width = 5, repr.plot.height = 5)

ggplot(pca_data, aes(x = PC1, y = PC2, label = SampleID,color=cohort.cohortGuid)) +
  geom_point(size = 3, color = "blue") +
  geom_text(vjust = -1, size = 3) +
  labs(title = "PCA of Scaled Cell Subset Frequencies",
       x = paste0("PC1 (", round(summary(pca_result)$importance[2, 1] * 100, 2), "% variance)"),
       y = paste0("PC2 (", round(summary(pca_result)$importance[2, 2] * 100, 2), "% variance)")) +
  theme_minimal()
```
