```r
suppressPackageStartupMessages(library(dplyr))
suppressPackageStartupMessages(library(tidyr))
suppressPackageStartupMessages(library(tibble))
suppressPackageStartupMessages(library(ggplot2))
suppressPackageStartupMessages(library(stringr))
library(patchwork)
library(gridExtra)
is_dark_color <- function(hex) {
  rgb <- col2rgb(hex)
  
  luminance <- 0.299 * rgb[1,] + 0.587 * rgb[2,] + 0.114 * rgb[3,]
  if(luminance < 128){
  return('white')
  }else{return('black')}
}
```

```r
color_annotation <- read.csv('../Color_hex_codes/Cell_Type_Color_Annotation_and_Order.csv')
color_vector <- setNames(color_annotation$color, color_annotation$label)
```

# Read NMF Scores

```r
df<-read.csv('03_Extract_NMF_Scores/Th2_Tfh_Baseline_Samples_BRI.csv')
```

```r
df <- df %>% dplyr::mutate(`Age Group` = case_when(
  cohort.cohortGuid == "BR1" ~ "Young",
  cohort.cohortGuid == "BR2" ~ "Older",
  TRUE ~ NA_character_
))
df$`Age Group`<-factor(df$`Age Group`,levels=c("Young","Older"))
```

```r
options(repr.plot.width = 75/25.4, repr.plot.height = 35/25.4)

p<-ggplot(df %>% filter(NMF_Scores=='NMF6_Tfh'), 
       aes(sample.daysSinceFirstVisit, AverageScore,color=`Age Group`)) +
  geom_point(position = position_jitter(width = 0.2), size=0.8,stroke=0) +
scale_color_manual(values = c("Older"='#bf812d',  "Young"='#35978f')) +theme_classic(base_size=6)+
  geom_smooth(se=FALSE,linewidth=0.5)+facet_wrap(~AIFI_L3)+xlab('')+theme(strip.background = element_rect(linewidth = 0))+ylab('Tfh Scores')+xlab('')+

  theme(
    axis.text.x = element_blank()  # Remove x-axis labels
  )+
  scale_y_continuous(labels = scales::number_format(accuracy = 0.001)) 
g1 <- ggplot_gtable(ggplot_build(p))

strips <- which(grepl('strip-', g1$layout$name))

pal <- color_vector['CM CD4 T cell']
text_color <- sapply(color_vector['CM CD4 T cell'], is_dark_color)

for (i in seq_along(strips)) {
  k <- which(grepl('rect', g1$grobs[[strips[i]]]$grobs[[1]]$childrenOrder))
  l <- which(grepl('titleGrob', g1$grobs[[strips[i]]]$grobs[[1]]$childrenOrder))
  g1$grobs[[strips[i]]]$grobs[[1]]$children[[k]]$gp$fill <-  adjustcolor(pal[i], alpha.f = 0.8)
  g1$grobs[[strips[i]]]$grobs[[1]]$children[[l]]$children[[1]]$gp$col <- text_color[i]
}

plot(g1)
```

```r
p<-ggplot(df %>% filter(NMF_Scores=='NMF5_Th2'), 
       aes(sample.daysSinceFirstVisit, AverageScore,color=`Age Group`)) +
  geom_point(position = position_jitter(width = 0.2), size=0.8,stroke=0) +
scale_color_manual(values = c("Older"='#bf812d',  "Young"='#35978f')) +theme_classic(base_size=6)+
  geom_smooth(se=FALSE,linewidth=0.5)+facet_wrap(~AIFI_L3)+xlab('')+theme(strip.background = element_rect(linewidth = 0))+ylab('Th2 Scores')+xlab('Sample Draw Date since Frist Visit')

g2 <- ggplot_gtable(ggplot_build(p))

strips <- which(grepl('strip-', g2$layout$name))

pal <- color_vector['CM CD4 T cell']
text_color <- sapply(color_vector['CM CD4 T cell'], is_dark_color)

for (i in seq_along(strips)) {
  k <- which(grepl('rect', g2$grobs[[strips[i]]]$grobs[[1]]$childrenOrder))
  l <- which(grepl('titleGrob', g2$grobs[[strips[i]]]$grobs[[1]]$childrenOrder))
  g2$grobs[[strips[i]]]$grobs[[1]]$children[[k]]$gp$fill <-  adjustcolor(pal[i], alpha.f = 0.8)
  g2$grobs[[strips[i]]]$grobs[[1]]$children[[l]]$children[[1]]$gp$col <- text_color[i]
}

plot(g2)
```

```r

options(repr.plot.width = 75/25.4, repr.plot.height = 70/25.4)

p1<-grid.arrange(g1, g2, ncol = 1)
ggsave('Extended-FigureE.pdf',p1,width = 75/25.4,height = 70/25.4)
```
