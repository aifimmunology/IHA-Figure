```r
projHeme1_subset <- addIterativeLSI(projHeme1_subset, name = 'IterativeLSI', force = TRUE, varFeatures = 75000)
```

```r
projHeme1_subset <- addHarmony(
    ArchRProj = projHeme1_subset,
    reducedDims = "IterativeLSI",
    name = "Harmony",
    groupBy = "Batch")
```

```r
projHeme1_subset <- addUMAP(
    ArchRProj = projHeme1_subset, 
    reducedDims = "Harmony", 
    name = "UMAPHarmony", 
    nNeighbors = 50, 
    minDist = 0.5, 
    metric = "cosine"
)
```

```r
projHeme1_subset <- addClusters(
    input = projHeme1_subset,
    reducedDims = "Harmony",
    method = "Seurat",
    name = "Clusters",
    resolution = 0.8,force = TRUE)
```

```r
pesudo_time_df<-as.data.frame(reducedDims(sce)$UMAP[o, ])
pesudo_time_df$pesudo_time<-x[o]
pesudo_time_df$pesudo_time_1<-as.data.frame(pseudotime)["Lineage1"][o,]
#pesudo_time_df$pesudo_time_2<-as.data.frame(pseudotime)["Lineage2"][o,]
pesudo_time_df$curve.assignments<-curve.assignments[o]

pesudo_time_df$celltype_new<-projHeme1_subset@cellColData$celltype_new[o]

pesudo_time_df$barcodes<-colData(sce)$barcodes[o]
```

```r
pesudo_time_df<-left_join(pesudo_time_df,motif_df, by=c("barcodes"='X'))
pesudo_time_df <- pesudo_time_df %>%
  mutate(celltype_new = as.character(celltype_new),  
         celltype_new = ifelse(is.na(celltype_new), "Core naive CD4 T cell", celltype_new),
         celltype_new = factor(celltype_new))
```

```r
pesudo_time_df <- pesudo_time_df %>%
  mutate('Age_Group' = case_when(
    startsWith(Subject, "BR2") ~ "Older",
    TRUE ~ 'Childern' 
  ))
```
