```r
meta_data=read.csv("/home//workspace/private/bri_figure_all_files_test/jupyter/BRI_Figures_Final_V2/Dataset/scRNA_meta_data-2024-05-09.csv")
meta_data=read.csv("/home//workspace/private/bri_figure_all_files_test/jupyter/BRI_Figures_Final_V2/Dataset/scRNA_meta_data-2024-05-09.csv")
meta_data$sample.drawDate <- as.Date(meta_data$sample.drawDate)
 
meta_data <- meta_data %>%
  mutate(Flu_Year = case_when(
    sample.drawDate <= as.Date('2020-07-01') ~ '2019-2020',
    sample.drawDate > as.Date('2020-07-01') & sample.drawDate <= as.Date('2021-07-01') ~ '2020-2021',
    sample.drawDate > as.Date('2021-07-01') & sample.drawDate <= as.Date('2022-07-01') ~ '2021-2022',
    sample.drawDate > as.Date('2022-07-01') & sample.drawDate <= as.Date('2023-07-01') ~ '2022-2023',
    TRUE ~ "NA"
  ))

meta_data <- meta_data %>%
  mutate(Flu_Day = case_when(
    sample.visitName	%in% c("Flu Year 1 Day 0","Flu Year 2 Day 0","Immune Variation Day 0") ~ 'Day_0',
    sample.visitName	%in% c("Flu Year 1 Day 7","Flu Year 2 Day 7",'Immune Variation Day 7') ~ 'Day_7',
    sample.visitName	%in% c("Flu Year 1 Day 90","Flu Year 2 Day 90",'Immune Variation Day 90') ~ 'Day_90',
    sample.visitName	%in% c("Flu Year 1 Stand-Alone","Flu Year 2 Stand-Alone","Flu Year 3 Stand-Alone") ~ 'Stand-Alone',
    TRUE ~ sample.visitName
  ))

meta_data <- meta_data %>%
  mutate(Visit_Type = case_when(
    startsWith(sample.visitName, "Flu Year") ~ "FluYear",
    startsWith(sample.visitName, "Immune Variation") ~ "ImmVar",
    TRUE ~ sample.visitName  
  ))

meta_data <- meta_data %>%
  mutate(Year_N = case_when(
    startsWith(sample.visitName, "Flu Year 1") ~ "Year 1",
    startsWith(sample.visitName, "Flu Year 2") ~ "Year 2",
      startsWith(sample.visitName, "Flu Year 2") ~ "Year 3",
    startsWith(sample.visitName, "Immune Variation") ~ "ImmVar",
    TRUE ~ sample.visitName  
  ))

meta_data <- meta_data %>% 
  dplyr::group_by(Year_N, subject.subjectGuid, Visit_Type) %>% 
  dplyr::arrange(Flu_Day) %>% 
  dplyr::mutate(Flu_Year = first(Flu_Year))

meta_data$Visit_Flu_Year<-paste0(meta_data$Visit_Type,"_",meta_data$Flu_Year,"_",meta_data$Flu_Day)
meta_data_subset <- meta_data %>%
    filter(Visit_Type == "FluYear", Flu_Year %in% c("2020-2021", "2021-2022"), Flu_Day != "Stand-Alone") %>%
    group_by(subjectGuid, Flu_Year) %>%
    mutate(entry_count = n()) %>%
    filter(entry_count == 3) %>%

    ungroup()
```

# Calculate Freq Changes

```r
df_freq<-read.csv("/home//workspace/private/bri_figure_all_files_test/jupyter/BRI_Figures_Final_V2/Dataset/diha_AIFI_L3_frequencies_2024-05-05.csv")
df_freq_subset<-df_freq %>% filter(specimen.specimenGuid %in% meta_data_subset$pbmc_sample_id)
```

```r
df_freq_subset <- left_join(df_freq_subset, meta_data_subset[c("pbmc_sample_id", "Flu_Day", "Flu_Year")], by = c(specimen.specimenGuid = "pbmc_sample_id"))
```

```r
freq_changes <- df_freq_subset %>%
    group_by(AIFI_L3, Flu_Year, subject.subjectGuid) %>%
    select(subject.subjectGuid, AIFI_L3_clr, Flu_Day) %>%
    tidyr::pivot_wider(names_from = Flu_Day, values_from = AIFI_L3_clr) %>%
    mutate(Freq_Fold_Changes_D7_D0 = Day_7 - Day_0) %>%
    mutate(Freq_Fold_Changes_D90_D0 = Day_90 - Day_0) %>%
    select(subject.subjectGuid,AIFI_L3, Flu_Year, Freq_Fold_Changes_D7_D0,Freq_Fold_Changes_D90_D0)
```

```r
freq_changes<-left_join(freq_changes,df_response_score %>% filter(Assay=='Flu B/Phuket HA'),by=c("subject.subjectGuid"='subjectGuid',"Flu_Year"="Flu_Year"))

```

```r
grouped_data <- freq_changes  %>%
  group_by(AIFI_L3, Flu_Year) %>%
  group_split()
```

```r

results <- mclapply(grouped_data, function(df) {


  df<-df %>% filter(responder %in% c("low responder","high responder"))
  df_res <- wilcox_test(df, Freq_Fold_Changes_D7_D0 ~ responder)
  df_res$Flu_Year <- df$Flu_Year[1]
  df_res$AIFI_L3 <- df$AIFI_L3[1]
  low <- median(df$Freq_Fold_Changes_D7_D0[df$responder == "low responder"], na.rm = TRUE)
  high<- median(df$Freq_Fold_Changes_D7_D0[df$responder == "high responder"], na.rm = TRUE)
  df_res$diff_clr <- high - low
  return(df_res)
}, mc.cores = 30)
```

```r
results_filter<-results[!grepl("Error",results)]
```

```r
df_res<-bind_rows(results_filter)
```

```r
df_res%>% as.data.frame(.)  %>% filter(grepl("",AIFI_L3))%>% filter(p<0.05)%>% arrange(p)
```

# Split BR1 and BR2

```r
grouped_data <- freq_changes  %>%
  group_by(AIFI_L3, Flu_Year, cohort.cohortGuid	) %>%
  group_split()
```

```r
results <- mclapply(grouped_data, function(df) {
    df <- df %>%
        filter(responder %in% c("low responder", "high responder"))
    df_res <- wilcox_test(df, Freq_Fold_Changes_D7_D0 ~ responder)
    df_res$Flu_Year <- df$Flu_Year[1]
    df_res$AIFI_L3 <- df$AIFI_L3[1]
    low <- median(df$Freq_Fold_Changes_D7_D0[df$responder == "low responder"], na.rm = TRUE)
    high <- median(df$Freq_Fold_Changes_D7_D0[df$responder == "high responder"], na.rm = TRUE)
    df_res$diff_clr <- high - low
    df_res$cohort <- unique(df$cohort.cohortGuid	)[1]

    return(df_res)
}, mc.cores = 30)
```

```r
results_filter<-results[!grepl("Error",results)]
```

```r
df_res<-bind_rows(results_filter)
```
