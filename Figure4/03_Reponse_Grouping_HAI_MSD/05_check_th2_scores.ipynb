```r
library(ggplot2)
library(stats)
library(dplyr)
library(tidyverse)
library(rstatix)
library(ggpubr)
library(parallel)
```

```r
meta_data<-read.csv("/home//workspace/IHA_Review_Analysis/Other_Analysis/13_MSD_HAI_analysis/meta_data.csv")
meta_data_subset <- meta_data %>%
    filter(Visit_Type == "FluYear", Flu_Year %in% c("2020-2021", "2021-2022"), Flu_Day != "Stand-Alone") %>%
    group_by(subjectGuid, Flu_Year) %>%
    mutate(entry_count = n()) %>%
    filter(entry_count == 3) %>%

    ungroup()
```

# Th2 Score

```r
df_list<-mclapply(meta_data$pbmc_sample_id,function(x){
df<-read.csv(paste0("/home//workspace/private/bri_figure_all_files_test/jupyter/BRI_Figures_Final_V2/Figure5/02_NMF_CD4_T_cell_Projection/NMF_Score_BRI/",x,".csv"))
df$pbmc_sample_id=x
return(df)
},mc.cores=30)
```

```r
df_combined_NMF<-do.call(rbind,df_list)
df_combined_NMF<- pivot_longer(df_combined_NMF, cols = starts_with("NMF"), names_to = "NMF_Scores", values_to = "value")
df_combined_NMF<-df_combined_NMF %>% group_by(pbmc_sample_id,NMF_Scores,AIFI_L3) %>% summarise(AverageScore=mean(value))
```

```r
df_combined_NMF<-df_combined_NMF %>% filter(grepl("Th2",NMF_Scores),AIFI_L3=="CM CD4 T cell")
```

```r
df_combined_NMF$sample.sampleKitGuid<-paste0("KT",substr(df_combined_NMF$pbmc_sample_id,3,7))
```

```r
df_combined_NMF<-df_combined_NMF%>% filter(pbmc_sample_id %in% meta_data_subset$pbmc_sample_id)
df_combined_NMF<-left_join(df_combined_NMF,meta_data_subset %>% select(pbmc_sample_id,subjectGuid,Flu_Year,Flu_Day)) %>% ungroup()
```

# Responder Group

```r
response_group<-read.csv('lm_res.csv')
```

```r
colnames(response_group)
```

```r
response_group<-left_join(response_group,df_combined_NMF %>% filter(Flu_Day=="Day_0"),by=c("Flu_Year","subjectGuid"	))
```

```r
options(repr.plot.width = 12, repr.plot.height = 15)

ggboxplot(
response_group %>% filter(responder
                          %in% c("low responder","high responder")),
    x = "responder", y = "AverageScore", color = "responder", palette = "jco",
    add = "jitter"
) + stat_compare_means() +
facet_wrap(~ Flu_Year +Assay , ncol = 4) + theme(axis.text.x = element_text(
    angle = 15, vjust = 0.5,
    hjust = 1
))+xlab("")
```

```r
options(repr.plot.width = 12, repr.plot.height = 15)

ggboxplot(
response_group %>% filter(responder
                          %in% c("low responder","high responder")),
    x = "cohort.cohortGuid", y = "AverageScore", palette = "jco",
    add = "jitter"
) + stat_compare_means() +
facet_wrap(~ Flu_Year +Assay , ncol = 4) + theme(axis.text.x = element_text(
    angle = 15, vjust = 0.5,
    hjust = 1
))+xlab("")
```
